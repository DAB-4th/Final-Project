```python
import pandas as pd

file_path = "/content/globalterrorismdb_0617dist.csv"

df = pd.read_csv(file_path, encoding='latin-1', low_memory=False)

print(df.shape)  # 데이터 크기 확인
print(df.head())  # 상위 5개 데이터 출력
```

```python
print(df.columns)
```

```python
# 'iyear'가 1990 이상인 데이터만 필터링
df_1990 = df[df['iyear'] >= 1990]

# 필터링 후 데이터 크기 확인
print(df_1990.shape)  # 필터링된 데이터의 크기
print(df_1990.head())  # 상위 5개 데이터 출력
```

```python
# CSV로 저장
df_1990.to_csv("/content/globalterrorism_1990.csv", index=False, encoding='utf-8-sig')
```
